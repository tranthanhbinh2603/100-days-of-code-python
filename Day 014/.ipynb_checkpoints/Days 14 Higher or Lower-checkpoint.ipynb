```python
from art import logo, vs
from game_data import data
import random

print(logo)

randa = random.randint(0, len(data)-1)
print(f"Compare A: {data[randa]['name']}, a {data[randa]['description']}, from {data[randa]['country']}.")

print(vs)

randb = random.randint(0, len(data)-1)
while randa == randb:
    randb = random.randint(0, len(data)-1)
print(f"Against B: {data[randb]['name']}, a {data[randb]['description']}, from {data[randb]['country']}.")

score = 0

while True:
    correctans = "A" if data[randa]['follower_count'] > data[randb]['follower_count'] else "B"
    if str(input("Who has more followers? Type 'A' or 'B': ")).upper() != correctans:
        print(f"Sorry, that's wrong. Final score: {score}")
        break
    else:
        score += 1
        print(f"You are right! Current score: {score}")
        if data[randa]['follower_count'] > data[randb]['follower_count']:
            randb = random.randint(0, len(data)-1)
        else:
            randa = randb
            randb = random.randint(0, len(data)-1)
        while randa == randb:
            randb = random.randint(0, len(data)-1)
        print(f"Compare A: {data[randa]['name']}, a {data[randa]['description']}, from {data[randa]['country']}.")
        print(vs)
        print(f"Against B: {data[randb]['name']}, a {data[randb]['description']}, from {data[randb]['country']}.")
```
